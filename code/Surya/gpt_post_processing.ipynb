```python
import openai
import os
from concurrent.futures import ThreadPoolExecutor, as_completed

# Set your API key
openai.api_key = os.getenv("OPENAI_API_KEY")

# Read the prompt text
with open("/Users/gladyscallow/Library/CloudStorage/OneDrive-UniversityofCambridge/JESUS/CONGRUENCE ENGINE/DIGITAL TOOLS/OCR/Surya/Worralls 1933/GPT process/GPT PROMPT.txt", "r") as prompt_file:
    prompt_text = prompt_file.read()

# Function to process text with GPT
def process_text_with_gpt(text):
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": prompt_text},
            {"role": "user", "content": text},
        ],
        max_tokens=4000,
    )
    cleaned_text = response["choices"][0]["message"]["content"]
    return cleaned_text.strip()

# Function to extract and append CSV content
def extract_and_append_csv_content(file_content, output_file, header):
    # Find the index where the CSV header starts
    start_index = 0
    lines = file_content.splitlines()
    for i, line in enumerate(lines):
        if line.strip().startswith(header):
            start_index = i
            break

    # Collect only the lines that are part of the CSV data
    csv_lines = lines[start_index:]
    csv_content = ""
    for line in csv_lines:
        if line.strip() == "":
            continue
        parts = line.strip().split(',')
        if len(parts) >= 4:
            csv_content += line + "\n"

    # Append the CSV content to the output file
    with open(output_file, 'a') as output:
        output.write(csv_content.strip() + "\n")

# Function to process a single file
def process_file(input_path, output_file, header):
    try:
        with open(input_path, 'r') as file:
            text = file.read()

        # Clean the text using GPT
        cleaned_text = process_text_with_gpt(text)

        # Extract and append CSV content to the consolidated output file
        extract_and_append_csv_content(cleaned_text, output_file, header)
        print(f"Processed {os.path.basename(input_path)}")
    except Exception as e:
        print(f"Error processing {os.path.basename(input_path)}: {e}")

# Function to process all files in parallel
def process_files_in_parallel(input_directory, output_file, header, max_workers=5):
    # Get all text files in the input directory, sorted numerically
    files = sorted(
        [os.path.join(input_directory, filename) for filename in os.listdir(input_directory) if filename.endswith(".txt")],
        key=lambda x: int(os.path.splitext(os.path.basename(x))[0])
    )

    # Ensure the output file exists and is empty
    open(output_file, 'w').close()

    # Process files in parallel using ThreadPoolExecutor
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_file = {executor.submit(process_file, file, output_file, header): file for file in files}
        for future in as_completed(future_to_file):
            try:
                future.result()  # Wait for the result to handle any exceptions
            except Exception as e:
                print(f"An error occurred: {e}")

# Define input directory, output file, and header
input_directory = '/Users/gladyscallow/Library/CloudStorage/OneDrive-UniversityofCambridge/JESUS/CONGRUENCE ENGINE/DIGITAL TOOLS/OCR/Surya/Worralls 1933/GPT process/test_input'
output_file = '/Users/gladyscallow/Library/CloudStorage/OneDrive-UniversityofCambridge/JESUS/CONGRUENCE ENGINE/DIGITAL TOOLS/OCR/Surya/Worralls 1933/GPT process/consolidated_output.csv'
header = "Companies,Addresses,Company Types,Bracketed Information"

# Process files in parallel and consolidate output
process_files_in_parallel(input_directory, output_file, header, max_workers=5)
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.jobs.ac.uk/job/DLF251/research-assistant-associate-modelling-marine-calcification-mechanisms-fixed-term"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract fields
title = soup.find("div", class_="j-advert__title row-3").text.strip()
employer = soup.find("div", class_="j-advert__employer row-4").text.strip()
details = soup.find("div", class_="j-advert-details__container row-5").text.strip()
description = soup.find("div", class_="job-description").text.strip()

# Combine into a dictionary
job_data = {
    "title": title,
    "employer": employer,
    "details": details,
    "description": description
}
print(job_data)
```
